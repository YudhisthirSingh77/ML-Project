# Model Building

```python
df.rename(columns = {'Avg. Area Income':'Area_Income', 'Avg. Area House Age':'Area_House_Age',
                     'Avg. Area Number of Rooms':'Number_of_Rooms','Area Population':'Area_Population'}
          , inplace = True)
```

```python
feats = ['Area_Income','Area_House_Age','Number_of_Rooms','Area_Population']
target = 'Price'
x = df[feats]
y = df[target]
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x,y,test_size=0.25,random_state=101)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

```python
xtrain
```

```python
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
score = model.score(xtest,ytest)
print(score)
```

```python
score*100
```

```python
ypred = model.predict(xtest)
```

```python
r2_score(ytest,ypred)
```

```python
plt.scatter(ytest,ypred)
```

```python
ypred[0],ytest.values[0]
```

```python
print('MAE :',mean_absolute_error(ytest,ypred))
```
